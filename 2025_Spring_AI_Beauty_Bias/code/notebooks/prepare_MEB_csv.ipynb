# Imports

```python
import pandas as pd
import os
import numpy as np
```

# Load raw data

```python
raw_df = pd.read_csv('../data/generic_scores_all_2022.csv')
raw_df
```

# Drop unnecessary columns

```python
raw_df.drop(columns=[col for col in raw_df.columns if col not in ['mean', 'image']], inplace=True)
raw_df.rename(columns={'mean': 'label'}, inplace=True)
raw_df
```

# Check for all files

```python
files = [file for file in os.listdir('MEB_cropped_images')]
files.__len__()
```

```python
raw_df = raw_df[raw_df['image'].isin(files)]
raw_df
```

# Make gender and race columns

```python
for gender in os.listdir('MEB_images'):
    for race in os.listdir(f'MEB_images/{gender}'):
        for file in os.listdir(f'MEB_images/{gender}/{race}'):
            if file in raw_df['image'].values:
                raw_df.loc[raw_df['image'] == file, 'gender'] = gender
                raw_df.loc[raw_df['image'] == file, 'race'] = race
raw_df
```

# Scale label to 0-1

```python
raw_df['label'] = (raw_df['label'] - raw_df['label'].min())
raw_df['label'] = raw_df['label'] / raw_df['label'].max()
raw_df
```

# Save the cleaned data

```python
raw_df.to_csv('../data/MEB.csv', index=False)
```
