# Imports

```python
import pandas as pd
import os
import numpy as np
```

# Load raw data

```python
df = pd.read_csv('../data/fairface_label_train.csv')
df
```

# Drop unnecessary columns

```python
df.drop(columns=['age', 'service_test'], inplace=True)
df
```

# Tansform columns

```python
df['image'] = df['file'].apply(lambda x: x.split('/')[-1])
df.drop(columns=['file'], inplace=True)
df
```

```python
df['gender'] = df['gender'].map({'Male': 'male', 'Female': 'female'})
df
```

```python
df['race'] = df['race'].str.lower()
df
```

# Check for all files

```python
files = os.listdir('fairface_cropped_images')
files.__len__()
```

```python
df = df[df['image'].isin(files)]
df
```

# Save the cleaned data

```python
df.to_csv('../data/FairFace.csv', index=False)
```
